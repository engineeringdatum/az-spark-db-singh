###  Ways to Drop

```
df.drop('id').show(1)
```

```
df.drop('email','gender').show(1)
```

```
#Even if a column name is mentioned and it's not available in the DF, it'll be ignored
df.drop('email','nothing').show(1)
```

* **Drop duplicates** 
    * Based on all columns. It is also known as distinct.
    * We can use `distinct`, `drop_duplicates` or `dropDuplicates` for these scenarios.

```
df.show()
```

```
df.count()
```

```
df.distinct().count()
```

```
df.dropDuplicates(['id', 'amount_paid']).show()
```

* **Drop records**
    * when all column values are nulls.
    * Drop records any of the column value is null.
    * Drop records that have less than `thresh` non-null values.
    * Drop records when any of the column value or all column values are nulls for provided subset of columns.
    * We can use `df.na.drop` or `df.dropna` to take care of dealing with records having columns with null values.

```
df.na
```

```
df.na.drop('all').show()
```

```
df.na.drop(thresh = 2).show() #Row having at-least 2 non-Null values
```

```
df.na.drop(how = 'any', subset = ['id','customer_from']).show()
```
